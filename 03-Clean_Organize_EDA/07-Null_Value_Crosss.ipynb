**Load Dataset**

```python
df = pd.read_csv(folder_path+"/10-WEO_Data.csv")
display(df.head())
```

```python

# Select only numeric columns for NaN check (exclude ID and categorical columns)
data_columns = df.columns.difference(['WEO_Country_Code', 'Country', 'Advanced_Country', 'Year'])

# Count NaNs in each row (for each country-year)
df['NaN_count'] = df[data_columns].isna().sum(axis=1)

# Sum NaNs per year across all countries
nan_per_year = df.groupby('Year')['NaN_count'].sum().reset_index()

# Years with no NaNs (total NaNs == 0)
years_no_nan = nan_per_year[nan_per_year['NaN_count'] == 0]['Year'].tolist()

# Year(s) with minimum NaNs
min_nan_count = nan_per_year['NaN_count'].min()
years_min_nan = nan_per_year[nan_per_year['NaN_count'] == min_nan_count]['Year'].tolist()

print("Years with no missing values:", years_no_nan)
print("Years with minimum missing values:", years_min_nan, "Missing count:", min_nan_count)
```

```python
# Using pandas plot (bar plot)
nan_per_year.plot(kind='bar', x='Year', y='NaN_count', color='skyblue', figsize=(12, 6))

plt.title('Number of Missing Values per Year')
plt.xlabel('Year')
plt.ylabel('Count of Missing Values')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# columns that missing values
cols_with_na = df_2015.columns[df_2015.isnull().any()].tolist()
cols_with_na
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.linear_model import BayesianRidge

# Copy 
df_2015_b = df_2015.copy()

# Handle missing values 
imputer = IterativeImputer(estimator=BayesianRidge(), max_iter=20, random_state=42, verbose=2)
df_2015_b[data_columns] = imputer.fit_transform(df_2015_b[data_columns])
```
